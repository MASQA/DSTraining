```python
def Precision_at_n(df_ratings, n=3, threshold=5):
    rel_n = set(df_ratings[df_ratings.real_rating>=threshold]['itemID'])
    t=df_ratings.sort_values(by='rating', ascending=False)[0:n]
    ret_n=set(t['itemID'])
    return len(rel_n & ret_n)/n
```

```python
def Avg_Precision_at_n(df_ratings, n=10,threshold=5):
    avg_p=0
    rel_n = set(df_ratings[df_ratings.real_rating>=threshold]['itemID'])

    for i in range(1,n+1):
        t=df_ratings.sort_values(by='rating', ascending=False)[0:i]
        ret_n=set(t['itemID'])
        t2=df_ratings.sort_values(by='rating', ascending=False)[i-1:i]
        ret_n2=set(t2['itemID'])
        #print(ret_n2,ret_n)
        if len((ret_n2 & rel_n))>0:
            avg_p=avg_p+len(rel_n & ret_n)/n
    return avg_p/n
```

```python
# Основная метрика проекта
def MAP_at_n(df_ratings_all, n=10,threshold=5):
    users=df_ratings_all['userID'].unique()
    map_at_n=0
    for i in users:
        df_ratings=df_ratings_all[df_ratings_all.userID==i]
        map_at_n=map_at_n+Avg_Precision_at_n(df_ratings,n=n,threshold=threshold)
    return map_at_n/len(users)
```

```python
# @title Функции для подготовки данных к поиску корреляций
def find_group_in_data(df, column):
  '''Трансформация. Вспомогательная. Находим количество групп в дискретной фиче'''
  return df.groupby(column).count().index.shape[0]


def find_label_groups(df, column):
  '''Трансформация. Вспомогательная. Составляем список из значений дискретной фичи'''
  return df.groupby(column).count().index.to_list()

def assign_number_by_item(list_items, items_to_repalce):
  '''Трансформация. Вспомогательная. Вычисляем значение индекса значения дискретной фичи,
  в массиве значений фичи '''
  return list_items.index(items_to_repalce)


def repalce_values_in_descret_columns(df, column):
  '''Трансформация. Новая колонка.
  Лэйбл значения дискретной фичи меняем на его индекс в массиве значений фичи'''
  list_values_column = find_label_groups(df, column)
  df[column+'_d']=df[column].apply(lambda x: assign_number_by_item(list_values_column, x))

def share_codes_discrete_values_between_ds(df, column, list_values):
  '''Трансформация. Новая колонка.
  Лэйбл значения дискретной фичи меняем на значение индекса в пришедшем массиве значений фичи'''
  df[column+'_d']=df[column].apply(lambda x: assign_number_by_item(list_values, x))
```

```python
def algebr_substraction(A,B):
  """ Solving from https://ru.stackoverflow.com/questions/642636/Быстрое-вычитание-списков"""
  C = list(set(A) - set(B))
  return(C)
```

### Класс подключения данных

```python
class DS_harness(object):
  """docstring"""

  def __init__(self):
    """Constructor"""
    # self.path = build_path_to_data(folder) + file_name
    self.dict_column_ohe = dict()
    self.dict_column_le = dict()
    self.dict_column_le_values = dict()

  def load_data(self, ds):
    # breakpoint()
    # if type_source == 'csv':
    # self.ds = pd.read_csv(self.path, sep=delimiter, index_col = index_is_need)
    self.ds = ds
    self.ds.reset_index(inplace = True)
    self.list_start_columns_set = self.ds.columns

  def transphorm_column(self, column, custom_function, add_column_need):
    # breakpoint()
    if add_column_need:
      self.ds[column + '_t'] = self.ds[column].apply(custom_function)
    else:
      self.ds[column] = self.ds[column].apply(custom_function)

  def build_list_columns_to_ohe(self, list_columns):
    self.list_columns_to_ohe = list_columns

  def code_column_ohe(self, column):
    # breakpoint()
    if list(self.dict_column_ohe.keys()).count(column) > 0:
      return f'For column {column} encoding has carried out already'
    enc = OneHotEncoder(handle_unknown='ignore',dtype = int)
    enc_df = pd.DataFrame(enc.fit_transform(self.ds[[column]]).toarray())
    # merge with main df bridge_df on key values
    enc_df.columns = [column + '_' + category for category in enc.categories_]
    enc_df.columns = [column[0].replace(' ','_')  for column in enc_df.columns]
    self.dict_column_ohe[column] = enc_df.columns
    self.ds = pd.concat([self.ds,enc_df],axis=1)

  def build_list_columns_to_le(self, list_columns):
    if list_columns is None:
      le_columns = []
      pattern = r'(_d)'
      for column in self.ds.columns:
        if re.search(pattern, column):
          le_columns.append(column.replace('_d',''))
          self.list_columns_to_le = le_columns
    else:
      self.list_columns_to_le = list_columns

  def code_column_le(self, column):
      if list(self.dict_column_le.keys()).count(column) > 0:
        return f'For column {column} encoding has carried out already'
      self.dict_column_le[column] = column + '_d'
      self.dict_column_le_values[column] = find_label_groups(self.ds, column)

  def get_list_ohe_columns_for_original(self, list_original_columns):
    list_ohe_columns = []
    list_not_encluded_columns = []
    for column in list_original_columns:
      if list(self.dict_column_ohe.keys()).count(column) > 0:
        list_ohe_columns.extend(self.dict_column_ohe[column].to_list())
      else:
        print(f'Column {column} is not coded by OneHotEncoder, column will be added as original')
        list_not_encluded_columns.append(column)
    return list_ohe_columns, list_not_encluded_columns

  def get_list_le_columns_for_original(self, list_original_columns):
    list_le_columns = []
    list_not_encluded_columns = []
    for column in list_original_columns:
      if list(self.dict_column_le.keys()).count(column) > 0:
        list_le_columns.extend([self.dict_column_le[column]])
      else:
        print(f'Column {column} is not coded by LabelEncoder, column will be added as original')
        list_not_encluded_columns.append(column)
    return list_le_columns, list_not_encluded_columns

  def build_list_features(self, column_map):
    # breakpoint()
    '''
        column_map = {
        target: <list columns target>,
        target_source: <le / ohe>
        exclude_columns: <list columns to_excluding>,
        feature_source: <le / ohe>
        list_additonal_columns: <list of additional column without coding>
      }
    '''
    list_columns = []
    list_additional_columns = []
    list_ohe_columns = []
    list_le_columns = []
    if column_map['target_source'] == 'le' :
      list_columns.extend([item +'_d' for item in column_map['target']])
    if column_map['target_source'] == 'ohe':
      list_columns.extend((self.get_list_ohe_columns_for_original(column_map['target'])[0]))
    list_remaining_columns = algebr_substraction(self.list_start_columns_set, column_map['target'])
    list_remaining_columns = algebr_substraction(list_remaining_columns, column_map['exclude_columns'])
    # list_remaining_columns = algebr_substraction(list_remaining_columns, column_map['list_additonal_columns'])
    if column_map['feature_source'] == 'le' :
        self.build_list_columns_to_le(None)
        for column in self.list_columns_to_le:
          self.code_column_le(column)
        # print(self.dict_column_le)
        list_columns.extend([column + '_d' if list(self.dict_column_le.keys()).count(column) else column for column in list_remaining_columns])
    if column_map['feature_source'] == 'ohe':
      list_ohe_columns, list_additional_columns = self.get_list_ohe_columns_for_original(list_remaining_columns)
      list_columns.extend(list_ohe_columns)
    list_columns.extend(column_map['list_additonal_columns'])
    return list_columns
```
