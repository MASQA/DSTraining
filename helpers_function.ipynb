```python

# не юзаю, отправить в набор функций для работы
# Конкатенация двух колонок года и месяца и любых дргуих с проставкой нуля для одинарных номеров месяца если дата и просот пробела если нуна строка 
def two_column_in_str(columns, delimiter, date = false):
  tmp = []
  for i in range(len(columns[0])):
    # print(str(columns[0][i]) + '_' +str(columns[1][i]))
    if (date) 
      if columns[1][i] // 10 == 0:
        tmp.append(str(columns[0][i]) + delimiter + '0' +str(columns[1][i]))
    else:
      tmp.append(str(columns[0][i]) + delimiter +str(columns[1][i]))
  return tmp
```

```python
# функция для определение метки даты 
# 'S' - если дельта между датами мешьше месяца
# 'M' - если дельта между датами не мешьше месяца и обе даты внутри одного года
# 'Y' - если дельта между датами не мешьше месяца и года отличаются
# 'ND' - если дельта между датами неможет быть рассчитана, так как даты окончания в выборке не представлено
# надо бы добавить уровень детализации день, месяц, год
def month_mention (x):
  if x.yearStartDate == x.yearEndDate and x.monthStartDate == x.monthEndDate:
    return 'S' 
  elif x.yearStartDate == x.yearEndDate and x.monthStartDate != x.monthEndDate:
    return 'M' 
  elif x.yearStartDate != x.yearEndDate and x.yearEndDate  != 'No Data':
    return 'Y' 
  else:
    return 'ND' 
```

```python
# Размница между датами в Часах
def diff_hour (x):
  d0 = x[0][0:19]
  d1 = x[1][0:19]
  format = '%Y-%m-%d %H:%M:%S'
  delta = abs(datetime.strptime(d1, format) - datetime.strptime(d0, format))
  # print(delta.days)
  return(divmod(delta.total_seconds(),3600))[0]
```

```python
# Размница между датами в Часах
def diff_day (x):
  d0 = x[0][0:19]
  d1 = x[1][0:19]
  format = '%Y-%m-%d %H:%M:%S'
  delta = abs(datetime.strptime(d1, format) - datetime.strptime(d0, format))
  # print(delta.days)
  return(divmod(delta.total_seconds(),86400))[0]
```

```python
# прическа легенды графиков

def legend_without_duplicate_labels(ax):
    handles, labels = ax.get_legend_handles_labels()
    unique = [(h, l) for i, (h, l) in enumerate(zip(handles, labels)) if l not in labels[:i]]
    ax.legend(*zip(*unique),fontsize =16 )
```

```python
# Так как размерность данных по некоторым курсам отличается, не находятся данные за определнный месяц,
# необходимо будет выравнивать  значения по ним
def normalizedListOfMedians(comparedList, normalizedData, etalon):
  # print(comparedList)
  # print(normalizedData)
  # print(etalon)
  resultList = []
  i = j = 0
  while i < len(etalon):
    if j >= len(normalizedData):
      resultList.append(0)
      i = i + 1
      continue
    if etalon[i] == comparedList [j]:
      resultList.append(normalizedData[j])
      i = i + 1
      j = j + 1
    else:
      resultList.append(0)
      i = i + 1
  # print(resultList)
  return resultList
```
